```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def get_image_name():
  path, dirs, files = next(os.walk(result_dir))
  file_count = len(files)
  return str(file_count + 1) + '.jpg'
```

```python
def crop_image(file_name, boundary_box):
    img = cv2.imread(os.path.join(data_dir,file_name))
    # print(os.path.join(data_dir,file_name))
    height = img.shape[0]
    width = img.shape[1]
    x=round(boundary_box[0] * width)
    y=round(boundary_box[1] * height)
    w=round(x+boundary_box[2] * width)
    h=round(y+boundary_box[3] * height)
    # print(x,y,w,h)
    return img[y:h, x:w]
```

```python
# fig = plt.figure(figsize=(10, 7))
# rows = 5
# columns = 5
# count = 1
for instance in f['images']:
    image = instance['file']
    detections = instance['detections']
        
        
    for detection in detections:
        print(detection['category'], detection['conf'])
        if detection['category'] == '1' and detection['conf'] >= 0.7:
            box = detection['bbox']
            print(image , "YES", box)
            cropped = crop_image(image,box)
            # fig.add_subplot(rows, columns, count)
            # plt.imshow(cropped)
            # count += 1
            cv2.imwrite(os.path.join(result_dir,image), cropped)
            # print(image, "NO")        
```
